```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
import pickle
```

```python
def get_translation(soup):
    return soup.find('div'
          ,attrs={'id':'translation'}
         ).text.strip()
```

```python
def get_next_path(soup):
    verse_sec = soup.find('div',attrs={'class':'verseNav'})
    next_url = verse_sec.find('div', attrs = {'class':'nextArrow'})
    elem = next_url.find('a')
    next_path = elem.get('href')
    return next_path
```

```python
def verse_content(soup):
    verses = soup.find('div',attrs={'lang':'sa','id':'originalVerse'}).find_all('p')
    verse_context = list()
    for verse in verses:
        verse_context.append(str(verse).replace('<br/>','\n').replace('</p>','').replace('<p>',''))
    return verse_context
```

```python
dict_verse = {}
dict_translation = {}
```

```python
def change_chapter(chapter):
    return f'https://www.holy-bhagavad-gita.org/chapter/{chapter}/verse/1'
```

```python
BASE_URL = 'https://www.holy-bhagavad-gita.org'
CHAPTER = 1
CUSTOM_URL = f'https://www.holy-bhagavad-gita.org/chapter/{CHAPTER}/verse/1'

key = CUSTOM_URL.replace(BASE_URL,'')
while True:
    try:
        r = requests.get(CUSTOM_URL)
        print(CUSTOM_URL)
        soup = BeautifulSoup(r.content, 'html5lib')
        next_path = get_next_path(soup)
        dict_verse[key]=verse_content(soup)
        dict_translation[key] = get_translation(soup)
        key = next_path
        CUSTOM_URL = BASE_URL+next_path
    except:
        CHAPTER+=1
        if CHAPTER==19:
            break
        CUSTOM_URL = change_chapter(CHAPTER)
```

```python
def get_hindi_url(chapter,verse):
    return f"https://www.gitasupersite.iitk.ac.in/srimad?language=dv&field_chapter_value={chapter}&field_nsutra_value={verse}&httyn=1&choose=1"
```

```python
dict_translation_hindi = dict()
```

```python
for chapter in tqdm(dict_chapter_verses):
    for verse in range(1,int(dict_chapter_verses[chapter])+1):
        CUSTOM_URL = get_hindi_url(chapter,verse)
        r = requests.get(CUSTOM_URL)
        soup = BeautifulSoup(r.content, 'html5lib')
        output = soup.find('div',attrs={'class':'views-field views-field-field-httyn'}).find_all('p')[-1].text
        dict_translation_hindi[f"/chapter/{chapter}/verse/{verse}"] = output
```
